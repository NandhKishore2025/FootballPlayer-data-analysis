```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("fifa_players.csv")
df.head(50)
```

```python
df.info()
```

```python
top_ten=df.sort_values(by='overall_rating', ascending=False)[['name','age',"stamina","strength","heading_accuracy",'finishing','overall_rating','long_shots','national_rating','weak_foot(1-5)']].head(10)
print(top_ten)
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(17, 6))
plt.scatter(top_ten['age'], top_ten['overall_rating'], color='green', s=100)
for i in range(len(top_ten)):
    plt.text(top_ten['age'].iloc[i] + 0.1, top_ten['overall_rating'].iloc[i],
             top_ten['name'].iloc[i], fontsize=9)
plt.title('Top 10 Players: Age vs Overall Rating')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_ten = df.sort_values(by='overall_rating', ascending=False)[['name', 'age', 'heading_accuracy', 'overall_rating']].dropna(subset=['age', 'heading_accuracy']).head(10)


plt.figure(figsize=(17, 6))
plt.scatter(top_ten['age'], top_ten['heading_accuracy'], color='blue', s=100)
for i in range(len(top_ten)):
    plt.text(top_ten['age'].iloc[i] + 0.1, top_ten['heading_accuracy'].iloc[i],top_ten['name'].iloc[i], fontsize=9)
plt.title('Top 10 Players: Age vs Heading_accuracy')
plt.xlabel('Age')
plt.ylabel('Heading_accuracy')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python



plt.figure(figsize=(17, 6))
plt.scatter(top_ten['age'], top_ten['national_rating'], color='purple', s=100)
for i in range(len(top_ten)):
    plt.text(top_ten['age'].iloc[i] + 0.2, top_ten['national_rating'].iloc[i], top_ten['name'].iloc[i], fontsize=9)
plt.title('Top 10 Players: Age vs National rating')
plt.xlabel('Age')
plt.ylabel('National rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```
